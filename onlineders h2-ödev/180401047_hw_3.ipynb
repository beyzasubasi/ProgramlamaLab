```python
#kütüphaneleri import ettik
from sympy import Symbol,pprint 
import sympy as sym  
import matplotlib.pyplot as plt
          
```

```python
# x ve lambdanın tanımlanması
x = Symbol('x')
lamd = Symbol('lambda')
```

```python
#exponential distribution formülü:1-(1-exp)=exp 
exponential_distribution = exp((-lamd)*x) 

#fonksiyonun günlük hayattaki yazımı
pprint(exponential_distribution)
```

```python
#üstel dağılımını yaptık,lambda 0.13; x, 0 ve 25 aralığında
sym.plot(exponential_distribution.subs({lamd:0.13}),(x,0,25),title='Exp Dist')

#boş dizi
x_values = []
y_values = []

for value in range(0,26):
    
    #formüle göre lambdayı 0.13 alarak istediğimiz değeri elde ettik
    y = exponential_distribution.subs({lamd:0.13, x: value}).evalf()
    x_values.append(value)
    y_values.append(y)
    
#grafiğimizi çizdirdik        
plt.plot(x_values,y_values)
plt.show()
```
